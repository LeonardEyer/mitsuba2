```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 1000)
y = (0.5 - np.arccos(x) / np.pi) * 2
```

```python
plt.plot(x, y)
```

```python
def smooth_profile(x, blur_size=0.5):
    x = np.array(x)
    res = np.zeros(x.shape)
    res[x >= blur_size] = 1.0
    res[x < blur_size] = (x[x < blur_size] / blur_size)**2
    return res;
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
y = np.cos(x)
y2 = (0.5 - np.arccos(y) / np.pi) * 2
plt.plot(x, y, label='cos')
plt.plot(x, y2, label='arccos')
plt.plot(x, smooth_profile(y2, .5), label='smooth')
plt.legend()
plt.ylim(-0.5, 1.5)
```
