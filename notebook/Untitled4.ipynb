```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
film_size = [5, 4]
spp = 100

total_sample_count = np.prod(film_size) * spp

pos = np.arange(total_sample_count)
idx = ((pos // (total_sample_count / film_size[1])) % film_size[1]) / film_size[1]
plt.plot(idx)
```
