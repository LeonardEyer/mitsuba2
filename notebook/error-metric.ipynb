```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate some test data
x = np.linspace(0, 100, 1000)
y = np.exp(- (x / 5))
y2 = np.exp(- (x / 10))
noise_decay = (np.log(x + 1) / 5)**10
noise = np.random.uniform(0, .5, 1000) * noise_decay
y += noise
plt.plot(y)
plt.plot(y2)
```

```python
def db(x):
    return 10 * np.log(x) / np.log(10)
```

```python
# MSE
mse = np.sum(np.square(y - y2)) / len(x)
mse
```

```python
# Decaying MSE
mse_decay = np.sum(np.square(y - y2) * np.exp(-x/10)) / len(x)
mse_decay
```

```python
plt.plot(np.square(y - y2), label='mse')
plt.plot(np.square(y - y2) * np.exp(-x/20), label='mse_decay')
plt.legend()
```
