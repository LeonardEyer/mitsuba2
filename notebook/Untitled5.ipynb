```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1, 0, 100)
plt.plot(x)
```

```python
def smooth_profile(blur_size, x):
    res = np.zeros(shape=(x.shape))
    res[x >= blur_size] = 1
    res[x < blur_size] = x[x < blur_size] / blur_size
    return res
```

```python
plt.plot(smooth_profile(0.5, np.cos(np.linspace(0, 2*np.pi, 1000))))
```

```python
smooth_profile(0.1, np.array([-0.7]))
```
