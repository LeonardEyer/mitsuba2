```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
# Generate a signal
x = np.linspace(0, 1, 1000)
y = np.sin(x*30) * np.cos(x * 5)
plt.plot(y)
```

```python
# Add some noise
noise = np.random.uniform(-0.7, 0.7, size=len(x))
noisy = y + noise
plt.plot(noisy)
```

```python
# Resample with low sampling rate
resampled = signal.resample(noisy, 100)
plt.plot(resampled)
```
